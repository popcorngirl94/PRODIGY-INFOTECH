```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_df = pd.read_csv('Population.csv', skiprows=2, header=2)
population_df.head()
```

```python
population_df.info()
```

```python
population_df.isna().sum()
```

```python
population_df = population_df.dropna()
```

```python
population_df.isna().sum()
```

```python
population_df.duplicated().sum()
```

```python
print(population_df.dtypes)
```

```python
sns.displot(data=population_df, x='2021', kind='hist',aspect=2, height = 8)
plt.title('Distribution of Population in Year 2021')
plt.xlabel('Year 2021')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=population_df, x='Country Name',y='2021',aspect=4, height=10, kind='bar')
plt.title('Country wise Population in 2021')
plt.xlabel('Country')
plt.ylabel('Year-2021')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.set_style("whitegrid")

population_column = '2021'
plt.figure(figsize=(10, 6))
sns.histplot(data=population_df, x=population_column, bins=20, color='skyblue', edgecolor='black', kde=True)
plt.title('Distribution of Population in Year 2021', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```

```python
sns.set_style("whitegrid")

population_column = '2021'
plt.figure(figsize=(15, 8))
sns.barplot(data=population_df, x='Country Name', y=population_column, color='skyblue')
plt.title('Country-wise Population in 2021', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel(f'Population in {population_column}', fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  
plt.show()
```
